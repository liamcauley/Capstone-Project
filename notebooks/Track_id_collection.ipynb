```python
for index, row in one_hit_charts.iterrows():
    song_response = spot.search(search_term = row.song, music_type='track', limit = 2)

    try:

        id = song_response.get('tracks').get('items')[0].get('id')

        track_name = song_response.get('tracks').get('items')[0].get('name')

        if track_name != row.song:
            print('maybe does not match')

        one_hit_charts.loc[index, 'track_id'] = id
        print(f'{index}/{len(one_hit_charts)-1}')

    except:
        print('no matches')
        one_hit_charts.loc[index, 'track_id'] = 'noid'
        continue
    finally:
        one_hit_charts.to_csv(f'../data/chart_id.csv', index=False)
```

```python
import pandas as pd
import SpotifyAPI as spot
```

```python
hits_80s = pd.read_csv('../data/80s_charts.csv')
hits_90s = pd.read_csv('../data/90s_charts.csv')
hits_2000to2010 = pd.read_csv('../data/2000s_2010s_charts.csv')
```

```python
song_response = spot.search(search_term = 'Funkytown', music_type='track', limit = 2)
```

```python
spot.test_getkey()
```

```python
for index, row in hits_80s.iterrows():
    song_response = spot.search(search_term = row.song, music_type='track', limit = 2)

    try:

        id = song_response.get('tracks').get('items')[0].get('id')

        track_name = song_response.get('tracks').get('items')[0].get('name')

        if track_name != row.song:
            print('maybe does not match')

        hits_80s.loc[index, 'track_id'] = id
        print(f'{index}/{len(hits_80s)-1}')

    except:
        print('no matches')
        hits_80s.loc[index, 'track_id'] = 'noid'
        continue
    finally:
        hits_80s.to_csv(f'../data/80s_chart_id.csv', index=False)
```

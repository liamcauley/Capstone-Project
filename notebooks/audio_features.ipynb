```python
import pandas as pd
import SpotifyAPI as spot

chart_id_2000 = pd.read_csv('../data/2000_2010s_chart_id.csv')


af_2000s = pd.DataFrame()

for index, row in chart_id_2000.iterrows():
    try:
        feature_response = spot.audio_features(track_id = row.track_id)
        
        temprow = pd.DataFrame(feature_response, index = [index])

        af_2000s = pd.concat([af_2000s, temprow])

        print(f'PROGRESS: {index} / {len(chart_id_2000)}')
    except:
        print(f'something went wrong with id {row.track_id}')
        continue
    finally:
        af_2000s.to_csv(f'../data/2000_2010_af.csv', index=False)
```

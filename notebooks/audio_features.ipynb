```python
import pandas as pd
import spotAPIfix as spot

chart_id = pd.read_csv('../data/chart_id.csv')


audio_features = pd.DataFrame()

for index, row in chart_id.iterrows():
    try:
        feature_response = spot.audio_features(track_id = row.track_id)
        
        temprow = pd.DataFrame(feature_response, index = [index])

        audio_features = pd.concat([audio_features, temprow])

        print(f'PROGRESS: {index} / {len(chart_id)}')
    except:
        print(f'something went wrong with id {row.track_id}')
        continue
    finally:
        audio_features.to_csv(f'../data/audio_features.csv', index=False)
```
